### Neste projeto trabalharemos com o Teorema do Limite Central, utilizando uma distribuição do tipo LogNormal.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats
```

```python
#1-a
s=1 #parâmetro da lognormal
```

```python
#1-b - Gerando uma distribuição lognormal aleatória
lognorm = stats.lognorm.rvs(s, size=1000000, random_state=None)
```

```python
# Calculando a média e a variância da distribuição gerada acima
mu = np.mean(lognorm)
var = np.var(lognorm)
print('Média: ', mu)
print('Variância teórica: ', var)
```

### Para n=2

```python
#2-a,b,c,d - Calculando a média e a variância para diversas escolhas de amostras de 2 elementos
lista2=[]
for j in range (0,10000):
    escolha2=[]
    for i in (0,2):
        escolha2.append(np.random.choice(lognorm))
    mu2=np.mean(escolha2)
    lista2.append(mu2)
```

```python
#2-e
fig, axs = plt.subplots(figsize=(8, 5))
plt.hist(lista2, bins=50, normed=1, color="salmon")
plt.title('Médias amostrais para n=2')
plt.xlabel("Médias")
plt.ylabel("Frequência relativa")
plt.show()
print('Média: ', np.mean(lista2))
print('Variância: ', np.var(lista2))
```

## O TLC tem o objetivo de calcular teoricamente os resultados da função teórica. Quando simulamos no python, divergencias podem ocorrer, portanto comparamos os resultados para diferentes espaços amostrais e analisamos a proximidade deles. A média é praticamente a mesma em todos os casos, enquando a variância é diferente, pois o TLC nos fornece uma variância para o espaço amostral mínimo (n=1), portanto para comparar, precisamos dividir este valor pelo novo espaço amostral simulado (n=2,, n=4, n=20, n=50). Comparando as médias e as variâncias calculadas para cada amostra com os valores teóricos obtidos pelo algoritmo do python, percebemos que o TLC funciona no nosso exemplo.

```python
#5
fig, axs = plt.subplots(figsize=(8, 5))
plt.hist(X1, bins=50, normed=1, color="red")
plt.title('Primeiros valores sorteados')
plt.xlabel("Valores")
plt.ylabel("Frequência relativa")
plt.axis([0,12,0,0.7])
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(8, 5))
plt.hist(X9, bins=50, normed=1, color="gold")
plt.title('Nonos valores sorteados')
plt.xlabel("Valores")
plt.ylabel("Frequência relativa")
plt.axis([0,12,0,0.7])
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(8, 5))
t = np.linspace(0,20,10000)
plt.hist(X1, bins=50, normed=1, color="red")
plt.title('Comparando')
plt.xlabel("Valores")
plt.ylabel("Frequência relativa")
plt.hist(X9, bins=50, normed=1, color='gold', alpha=0.7)
plt.plot(t,stats.lognorm.pdf(t, s, loc=0, scale=1), color='blue')
plt.axis([0,12,0,0.7])
plt.show()
```

```python
cov=0
for i in range(len(X1)):
    cov+=(X1[i]-np.mean(X1))*(X9[i]-np.mean(X9))/(len(X1))
corr=cov/(np.var(X1)*np.var(X9))**0.5
print("Correlação entre X1 e X9: ", corr)
```
